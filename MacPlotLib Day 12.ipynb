```python
batsman= pd. read_csv("sharma-kohli.csv")
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

```python
# Labels and title

plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])

plt.title("Rohit Sharma vs Virat Kohli career comparison")
plt.xlabel("Season")
plt.ylabel("Run Record")
```

```python
# Colors/ Line width and style / marker size

plt.plot(batsman["index"],batsman["V Kohli"],color="Purple")
plt.plot(batsman["index"],batsman["RG Sharma"],color="Grey")

plt.title("Rohit Sharma vs Virat Kohli career comparison")
plt.xlabel("Season")
plt.ylabel("Run Record")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="Purple",linestyle="dashed", linewidth=3)
plt.plot(batsman["index"],batsman["RG Sharma"],color="Grey",linestyle="dashdot",linewidth=5)

plt.title("Rohit Sharma vs Virat Kohli career comparison")
plt.xlabel("Season")
plt.ylabel("Run Record")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="Purple",linestyle="dashed", linewidth=3,marker="o",markersize=10)
plt.plot(batsman["index"],batsman["RG Sharma"],color="Grey",linestyle="dashdot",linewidth=5,marker="+",markersize=10)

plt.title("Rohit Sharma vs Virat Kohli career comparison")
plt.xlabel("Season")
plt.ylabel("Run Record")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="Purple",linestyle="dashed", linewidth=3,marker="o",markersize=10, label="Virat Kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],color="Grey",linestyle="dashdot",linewidth=5,marker="+",markersize=10, label="Rohit Sharma")

plt.title("Rohit Sharma vs Virat Kohli career comparison")
plt.xlabel("Season")
plt.ylabel("Run Record")

plt.legend(loc="best")
```

```python
# Limiting Axis

price= [48000,54000,57000,49000,47000,45000,450000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,250000)
plt.xlim(2017,2019)
```

```python
# grid 

plt.plot(batsman["index"],batsman["V Kohli"],color="Purple",linestyle="dashed", linewidth=3,marker="o",markersize=10,label="Virat Kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],color="Grey",linestyle="dashdot",linewidth=5,marker="+",markersize=10,label="Rohit Sharma")

plt.title("Rohit Sharma vs Virat Kohli career comparison")
plt.xlabel("Season")
plt.ylabel("Run Record")
plt.legend(loc="best")
plt.grid()
plt.show()
```
